```python
import pandas as pd

df = pd.read_csv("../data/raw_dataset.csv")

df.head()
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
import matplotlib.pyplot as plt

sales = df.groupby("Category")["Sales"].sum()

sales.plot(kind="bar")

plt.title("Sales by Category")

plt.show()
```
